```python
def BernoulliTrail(p=0.5):
    X = int(np.random.rand()<=p)
    return X
```

```python
print(BernoulliTrail(0.7))
```

```python
count = 0
n = 1000000
for _ in range(n):
    if BernoulliTrail(0.1):
        count += 1
print(count/n)
```

```python
def GT(p):
    X = 1
    while True:
        if BernoulliTrail(p)==1:
            return X 
        else:
            X += 1
    
```

```python
p = 0.01
print(GT(p))
```

# Geometric(p), $E[X] = 1/p$

```python
# E[X] = 1/p
print(1/p)
p = 0.812
SmeanX = []
for N in range(1, 10000, 100):
    SmeanX.append((np.random.geometric(p, N).mean()))
SmeanX = np.array(SmeanX)
y = np.array(SmeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```

# Binomial(n,p) $E[X] =np$

```python
# E[X] = np
n = 350
print(n*p)
p = 0.812
SmeanX = []
for N in range(1, 10000, 100):
    SmeanX.append((np.random.binomial(n,p, N).mean()))
SmeanX = np.array(SmeanX)
y = np.array(SmeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```

# Normal($\mu, \sigma $), $E[X] =\mu$

```python
# E[X] = mu
mu = 2.4
sgma = 3.1
SmeanX = []
for N in range(1, 10000, 100):
    SmeanX.append((np.random.normal(mu, sgma, N).mean()))
SmeanX = np.array(SmeanX)
y = np.array(SmeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```
